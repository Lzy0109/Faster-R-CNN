## TP、TN、FP、FN

TP：True Positives，**分配为正样本，且分配对了**的样本，代表的是**被正确分类的正样本（真正例）**。

TN：True Negative，**分配为负样本，且分配对了**的样本，代表的是**被正确分类的负样本（真反例）**。

FP：False Positives，**分配为正样本，但分配错了**的样本，代表的是**被错误分类的负样本（假正例）**。

FN：False Negatives，**分配为负样本，但分配错了**的样本，代表的是**被错误分类的正样本（假反例）**。

Precision：精度，**分类器认为是正类，且确实为正类部分**占**分类器认为是正类**的比例。

$Precision=\frac{TP}{TP+FP}$

Recall：**分类器认为是正类且确实是正类的部分**占**所有确实是正类**的比例。

$Recall=\frac{TP}{TP+FN}$

目标检测任务中，**每一个类都可以计算出其Precision和Recall**，通过合理的计算，每个类都可以**得到一条P-R曲线**，**曲线下的面积就是AP的值。**


## mAP
假设存在**M张图片**，对于其中一张图片而言，其具有**N个检测目标**，其具有**K个类**，使用检测器得到了**S个Bounding Box**(BB)，每个BB里包含BB所在的位置以及对于K个类的**得分C**。利用BB所在的位置可以得到与其对应的GroundTruth的**IOU值**。

*   对于**每一个类I**，计算**BB所在位置与其最对应的GroundTruth的IOU值，记MaxIOU。**与阈值threshold比较，通常为0.5。
*   **若MaxIOU < threshold**, 认为预测框与真实框无对应，属于False Positive。令其$FP_i=1$，记录其属于类I的分数。
*   **若MaxIOU > threshold**，认为预测框与真实框对应，再分类：
  
    1.   若属于类I，记为True Positive，令其$TP_i=1$,记录属于类I的分数。
    2.   若不属于，记为False Positive，令其$FP_i=1$,记录属于类I的分数。

*   最终获得K * S个分数C和TP、FP的元组，构成(C，TP，FP)的元组，按C进行从大到小排序。
*   截取得分最大的S个，**这个步骤可得知每个框是否成功对应自己所属的类。**计算每次截取的Precision和Recall，得到S个Recall点和Precision点。
*   所连成的曲线即为**P-R曲线。**PR曲线下面的面积即为AP的值，mAP是对所有AP值进行求平均。
